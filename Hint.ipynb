```python
import pandas as pd

data = pd.read_csv("/tmp/hasil_arbi.csv")

data = data.sort_values(by=data.columns[3], ignore_index=True)

data = data.iloc[:, 1:]

icols = ["Product ID", "Total", "Support"]

data.columns = icols

data.iloc[:5]
```

```python
data[data[data.columns[0]].str.contains("\n")]
```

```python
import re

data[icols[0]] = data[icols[0]].apply(lambda i : re.sub("\n", "", i))

data.head()
```

```python
data = data.drop_duplicates(keep="first")

data.shape
```

```python
data = data[data[icols[1]].isin([3, 4, 5])]

data.shape
```

```python
data = data.reset_index(drop=True)

data.iloc[:5]
```

```python
product_df = []

pt = data.iloc[:, 0]

for i in range(len(pt)):
  for t in range(i + 1):
    if pt[i] == pt[t]:
      continue
    product_df.append(" ".join([pt[i], pt[t]]))
```
